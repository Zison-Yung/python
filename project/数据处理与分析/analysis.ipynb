# 数据分析

```python
import numpy as np
import pandas as pd
import re
df = pd.read_csv(r'rent_source_合并.csv', header= 0,
                 names =['title','location','price','url','source'] )
```

```python
df[:5]
```

```python
#新增一列，作为区域地址
df['address'] = 'add'
for index in range(len(df)):
    df.loc[index,'address'] = re.findall('(.*?) .*',df.loc[index,'title'])
```

```python
df[:10]
```

```python
#总体最低价
min = df['price'].idxmin()
df.iloc[min]
```

```python
#总体最高价
min = df['price'].idxmax()
df.iloc[min]
```

```python
#总体均价
df['price'].mean()
```

```python
#总体描述性统计
df.describe()
```

```python
#各区域房源数量
df['address'].value_counts()
```

```python
#各地区均价和最低价
pvt = pd.pivot_table(df,index=['address'],values=['price'],aggfunc=[np.mean,np.min])
pvt.sort_values(by=('mean','price'))
```

```python
#离散化处理
df['price_grade'] = pd.cut(df.price,
                           bins=[0,1500,3000,999999999],
                           labels=['低','中','高']);
df_grp = df['price_grade'].groupby(df["price_grade"]).count()
df_grp
```
